```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
import scipy.stats as sc
from scipy.stats import gamma
from scipy.stats import beta
from scipy.stats import chi2
%matplotlib inline

fig, axes = plt.subplots(7,2,figsize=(12, 32));

D = ['r','g','b'];

#######################################################################################################################
#######################################################################################################################


## Uniform Distribution


a = [-5,-1,2];
b = [-2,1,6];
L = ['a= -5,b= -2','a= -1,b= -1','a= 2,b= 6']

## Probability Distribution function of Uniform Distribution

for i in range(0,3):
    y = np.linspace(a[i]-1,b[i]+1,1000);

    def uniform_pdf(x,a,b):
        if a[i] <= x <= b[i]:
            return 1/(b[i]-a[i]);
        else:
            return 0;
        
    mean = (a[i]+b[i])/2;
    median = (a[i]+b[i])/2;
    mode = random.uniform(a[i],b[i]);
    variance = ((b[i]-a[i])**2)/12;
    
    axes[0,0].plot(y,[uniform_pdf(x,a,b) for x in y],D[i],label=L[i]);
    axes[0,0].plot(mean,1/(b[i]-a[i]),'ko');
    axes[0,0].plot(median,1/(b[i]-a[i]),'y.');
    axes[0,0].plot(mode,1/(b[i]-a[i]),'r*');
    axes[0,0].plot(variance,1/(b[i]-a[i]),'ks');
    
    
axes[0,0].plot(mean,1/4,'ko',label='Mean');
axes[0,0].plot(median,1/4,'y.',label='Median');
axes[0,0].plot(mode,1/4,'r*',label='Mode');
axes[0,0].plot(variance,1/4,'ks',label='Variance');
axes[0,0].axhline(color = "Black");
axes[0,0].axvline(color = "Black");
axes[0,0].set_xlabel("x - axis");
axes[0,0].set_ylabel("Probability Distribution");
axes[0,0].set_title("PDF Of Uniform Distribution");
axes[0,0].legend()


## Cumulative Distribution function of Uniform Distribution

for i in range(0,3):
    y = np.linspace(a[i]-1,b[i]+1,1000);

    def uniform_cdf(x,a,b):
        if x < a[i]:
            return 0;
        elif a[i] <= x <= b[i]:
            return (x - a[i])/(b[i] - a[i]);
        elif x >= b[i]:
            return 1;

    axes[0,1].plot(y,[uniform_cdf(x,a,b) for x in y],D[i],label=L[i]);
    
axes[0,1].axhline(color = "Black");
axes[0,1].axvline(color = "Black");
axes[0,1].set_xlabel("x - axis");
axes[0,1].set_ylabel("Cumulative Distribution");
axes[0,1].set_title("CDF Of Uniform Distribution")
axes[0,1].legend()
    
    

#######################################################################################################################
#######################################################################################################################
  


## Exponential Distribution


lamda = [0.5,1,1.5];
L = ['λ = 0.5','λ = 1','λ = 1.5']

## Probability Distribution function of Exponential Distribution
    
for i in range(0,3):
    
    x = np.linspace(0,5);
    
    def exponential_pdf(x,lamda):
        f1 = lamda * np.exp(-lamda * x);
        return f1;

    mean = 1/lamda[i];
    median = np.log(2)/lamda[i]
    variance = 1/(lamda[i])**2;
    
    axes[1,0].plot(x,exponential_pdf(x,lamda[i]),D[i],label=L[i]);
    axes[1,0].plot(variance,exponential_pdf(variance,lamda[i]),'bs');
    axes[1,0].plot(median,exponential_pdf(median,lamda[i]),'r*');
    axes[1,0].plot(mean,exponential_pdf(mean,lamda[i]),'ko');
    
    
axes[1,0].plot(mean,exponential_pdf(mean,1),'ko',label='Mean');
axes[1,0].plot(median,exponential_pdf(median,1),'r*',label='Median');
axes[1,0].plot(variance,exponential_pdf(variance,1),'bs',label='Variance');
axes[1,0].axhline(color = "Black");
axes[1,0].axvline(color = "Black");
axes[1,0].set_xlabel("x - axis");
axes[1,0].set_ylabel("Probability Distribution");
axes[1,0].set_title("PDF Of Exponential Distribution")
axes[1,0].legend()
    
    
## Cumulative Distribution function of Exponential Distribution
    
    
for i in range(0,3):
    
    x = np.linspace(0,5);

    def exponential_cdf(x,lamda):
        f2 = 1 - np.exp(-lamda * x);
        return f2;

    axes[1,1].plot(x,exponential_cdf(x,(lamda[i])),D[i],label=L[i]);

    
axes[1,1].axhline(color = "Black");
axes[1,1].axvline(color = "Black");
axes[1,1].set_xlabel("x - axis");
axes[1,1].set_ylabel("Cumulative Distribution");
axes[1,1].set_title("CDF Of Exponential Distribution");
axes[1,1].legend(loc=(0.795,0.05))
    
    
    
#######################################################################################################################
#######################################################################################################################


## Gaussian Distribution

mu = [0,-2,0];
sigma = [0.45,0.707,2.236];
L = ['μ = 0,σ = 0.45','μ = -2,σ = 0.707','μ = 0,σ = 2.236']

## Probability Distribution function of Gaussian Distribution


for i in range(0,3):
    x = np.linspace(-5,5);

    def gaussian_pdf(x,mu,sigma):
        f1 = (((x - mu)/sigma)**2)/2;
        f2 = 1/(sigma*((2*np.pi)**0.5));
        f3 = f2 * np.exp(-f1);
        return f3;
    
    mean = mu[i];
    median = mu[i];
    mode = mu[i];
    variance = (sigma[i])**2;
    
    axes[2,0].plot(x,gaussian_pdf(x,mu[i],sigma[i]),D[i],label=L[i]);
    axes[2,0].plot(mean,gaussian_pdf(mean,mu[i],sigma[i]),'ko');
    axes[2,0].plot(mode,gaussian_pdf(mode,mu[i],sigma[i]),'y>');
    axes[2,0].plot(median,gaussian_pdf(median,mu[i],sigma[i]),'r.');
    axes[2,0].plot(variance,gaussian_pdf(variance,mu[i],sigma[i]),'ks');
    
    
axes[2,0].plot(mean,gaussian_pdf(mean,0,0.45),'ko',label='Mean');
axes[2,0].plot(median,gaussian_pdf(median,0,0.45),'r.',label='Median');
axes[2,0].plot(mode,gaussian_pdf(mode,0,0.45),'y>',label='Mode');
axes[2,0].plot(variance,gaussian_pdf(variance,0,0.45),'ks',label='Variance');
axes[2,0].axhline(color = "Black");
axes[2,0].axvline(color = "Black");
axes[2,0].set_xlabel("x - axis");
axes[2,0].set_ylabel("Probability Distribution");
axes[2,0].set_title("PDF Of Gaussian Distribution")
axes[2,0].legend()


## Cumulative Distribution function of Gaussian Distribution


F = [];

for i in range(0,3):
    
    x = np.linspace(-5,5,1000);

    def gaussian_cdf(x,mu,sigma):
        for j in range(0,1000):
            f = (1 + (math.erf((x[j] - mu)/(sigma*1.414))))*0.5;
            F.append(f);
        return F;
    
    F = [];
    
    axes[2,1].plot(x,gaussian_cdf(x,mu[i],sigma[i]),D[i],label=L[i]);
    
axes[2,1].axhline(color = "Black");
axes[2,1].axvline(color = "Black");
axes[2,1].set_xlabel("x - axis");
axes[2,1].set_ylabel("Cumulative Distribution");
axes[2,1].set_title("CDF Of Gaussian Distribution")
axes[2,1].legend()
    
    
    
#######################################################################################################################
#######################################################################################################################



## Standard Normal Distribution

mu = [-1,0,1];
sigma = 1;
L = ['μ = -1,σ = 1','μ = 0,σ = 0','μ = 1,σ = 1'];

## Probability Distribution function of Standard Normal Distribution

for i in range(0,3):
    
    x = np.linspace(-5,5,1000);

    def standard_normal_pdf(x,mu,sigma):
        f1 = (((x - mu)/sigma)**2)/2;
        f2 = 1/(sigma*((2*np.pi)**0.5));
        f3 = f2 * np.exp(-f1);
        return f3;
    
    
    mean = mu[i];
    median = mu[i];
    mode = mu[i];
    variance = (sigma)**2;

    axes[3,0].plot(x,standard_normal_pdf(x,mu[i],sigma),D[i],label=L[i]);
    axes[3,0].plot(mean,standard_normal_pdf(mean,mu[i],sigma),'sb');
    axes[3,0].plot(mode,standard_normal_pdf(mode,mu[i],sigma),'or');
    axes[3,0].plot(median,standard_normal_pdf(median,mu[i],sigma),'k.');
    axes[3,0].plot(variance,standard_normal_pdf(variance,mu[i],sigma),'ko');

    
    
axes[3,0].plot(mean,standard_normal_pdf(mean,0,1),'sb',label='Mean');
axes[3,0].plot(median,standard_normal_pdf(median,0,1),'k.',label='Median');
axes[3,0].plot(mode,standard_normal_pdf(mode,0,1),'or',label='Mode');
axes[3,0].plot(variance,standard_normal_pdf(variance,0,1),'ko',label='Variance');
axes[3,0].axhline(color = "Black");
axes[3,0].axvline(color = "Black");
axes[3,0].set_xlabel("x - axis");
axes[3,0].set_ylabel("Probability Distribution");
axes[3,0].set_title("PDF Of Standard Normal Distribution")
axes[3,0].legend()


## Cumulative Distribution function of Standard Normal Distribution

F = [];

for i in range(0,3):
    
    x = np.linspace(-5,5,1000);

    def standard_normal_cdf(x,mu,sigma):
        for j in range(0,1000):
            f = (1 + (math.erf((x[j] - mu)/(sigma*1.414))))*0.5;
            F.append(f);
        return F;
    
    F = [];
    
    axes[3,1].plot(x,standard_normal_cdf(x,mu[i],sigma),D[i],label=L[i]);
    
axes[3,1].axhline(color = "Black");
axes[3,1].axvline(color = "Black");
axes[3,1].set_xlabel("x - axis");
axes[3,1].set_ylabel("Cumulative Distribution");
axes[3,1].set_title("CDF Of Standard Normal Distribution");
axes[3,1].legend()
    
    
    
#######################################################################################################################
#######################################################################################################################



## Gamma Distribution


k = [1,7,9];
theta = [2,1,0.5];
L = ['k = 1,θ = 2','k = 7,θ = 1','k = 9,θ = 0.5'];

## Probability Distribution function of Gamma Distribution

for i in range(0,3):
    x = np.linspace(0,18,1000)

    def gamma_pdf(x,k,theta):
        f1 = math.gamma(k);
        f2 = 1/(f1*(theta**k));
        f3 = f2*((x)**(k-1)) * np.exp(-(x/theta));
        return f3;
    
    mean = k[i]*theta[i]; 
    mode = (k[i]-1)*theta[i];
    variance = k[i]*((theta[i])**2);

    axes[4,0].plot(x,gamma_pdf(x,k[i],theta[i]),D[i],label=L[i]);
    axes[4,0].plot(mean,gamma_pdf(mean,k[i],theta[i]),'ro');
    axes[4,0].plot(mode,gamma_pdf(mode,k[i],theta[i]),'ks');
    axes[4,0].plot(variance,gamma_pdf(variance,k[i],theta[i]),'b*');

    
axes[4,0].plot(mean,gamma_pdf(mean,1,2),'ro',label='Mean');
axes[4,0].plot(mode,gamma_pdf(mode,1,2),'ks',label='Mode');
axes[4,0].plot(variance,gamma_pdf(variance,1,2),'b*',label='Variance');
axes[4,0].axhline(color = "Black");
axes[4,0].axvline(color = "Black");
axes[4,0].set_xlabel("x - axis");
axes[4,0].set_ylabel("Probability Distribution");
axes[4,0].set_title("PDF Of Gamma Distribution")
axes[4,0].legend()


## Cumulative Distribution function of Gamma Distribution
    
for i in range(0,3):
    
    x = np.linspace(0,18,1000)
    
    def gamma_cdf(x,k):
        f = gamma.cdf(x,k);
        return f;
    
    
    axes[4,1].plot(x,gamma_cdf(x,k[i]),D[i],label=L[i]);
    
axes[4,1].axhline(color = "Black");
axes[4,1].axvline(color = "Black");
axes[4,1].set_xlabel("x - axis");
axes[4,1].set_ylabel("Cumulative Distribution");
axes[4,1].set_title("CDF Of Gamma Distribution");
axes[4,1].legend(loc=(0.705,0.05))
    

#######################################################################################################################
#######################################################################################################################

    
## Beta Distribution

alpha = [1,2,2];
betha = [3,2,5];
L = ['α = 1,β = 3','α = 2,β = 2','α = 2,β = 5'];


## Probability Distribution function of Beta Distribution


for i in range(0,3):
    x = np.linspace(0,1,1000)

    def betha_pdf(x,alpha,betha):
        f1 = (math.gamma(alpha)*math.gamma(betha))/math.gamma(alpha+betha);
        f2 = x**(alpha - 1)
        f3 = (((1-x)**(betha-1))*f2)/f1;
        return f3;
    
    mean = alpha[i]/(alpha[i]+betha[i]);
    median = (alpha[i]-(1/3))/(alpha[i]+betha[i]-(2/3));
    mode = (alpha[i]-1)/(alpha[i]+betha[i]-2);
    variance = (alpha[i]*betha[i])/(((alpha[i]+betha[i])**2)*(alpha[i]+betha[i]+1));

    axes[5,0].plot(x,betha_pdf(x,alpha[i],betha[i]),D[i],label=L[i]);
    axes[5,0].plot(mean,betha_pdf(mean,alpha[i],betha[i]),'r*');
    axes[5,0].plot(median,betha_pdf(median,alpha[i],betha[i]),'ks');
    axes[5,0].plot(mode,betha_pdf(mode,alpha[i],betha[i]),'y.');
    axes[5,0].plot(variance,betha_pdf(variance,alpha[i],betha[i]),'ko');
    

axes[5,0].plot(mean,betha_pdf(mean,1,3),'r*',label='Mean');
axes[5,0].plot(median,betha_pdf(median,1,3),'ks',label='Median');
axes[5,0].plot(mode,betha_pdf(mode,1,3),'y.',label='Mode');
axes[5,0].plot(variance,betha_pdf(variance,1,3),'ko',label='Variance');
axes[5,0].axhline(color = "Black");
axes[5,0].axvline(color = "Black");
axes[5,0].set_xlabel("x - axis");
axes[5,0].set_ylabel("Probability Distribution");
axes[5,0].set_title("PDF Of Beta Distribution");
axes[5,0].legend()


## Cumulative Distribution function of Beta Distribution

    
for i in range(0,3):
    
    x = np.linspace(0,1,1000);
    
    def beta_cdf(x,alpha,betha):
        f = beta.cdf(x,alpha,betha);
        return f;
    
    
    axes[5,1].plot(x,beta_cdf(x,alpha[i],betha[i]),D[i],label=L[i]);
    
axes[5,1].axhline(color = "Black");
axes[5,1].axvline(color = "Black");
axes[5,1].set_xlabel("x - axis");
axes[5,1].set_ylabel("Cumulative Distribution");
axes[5,1].set_title("CDF Of Beta Distribution");
axes[5,1].legend(loc=(0.73,0.05))
    
    

#######################################################################################################################
#######################################################################################################################
    
    
    
## Chi Square Distribution

k = [2,6,9];
L = ['k = 2','k = 6','k = 9'];


## Probability Distribution function of Chi Squared Distribution


for i in range(0,len(k)):
    
    x = np.linspace(0,20,1000)

    def chi_square_pdf(x,k):
        f1 = math.gamma(k/2);
        f3 = 1/(((2)**(k/2))*f1);
        f4 =  ((x)**((k/2)-1))*(np.exp(-(x/2)))*f3
        return f4;
    
    mean = k[i];
    median = k[i]*((1-(2/(9*k[i])))**3);
    mode = max((k[i]-2),0);
    variance = 2*k[i];

    axes[6,0].plot(x,chi_square_pdf(x,k[i]),D[i],label=L[i]);
    axes[6,0].plot(mean,chi_square_pdf(mean,k[i]),'mo');
    axes[6,0].plot(median,chi_square_pdf(median,k[i]),'k*');
    axes[6,0].plot(mode,chi_square_pdf(mode,k[i]),'ks');
    axes[6,0].plot(variance,chi_square_pdf(variance,k[i]),'ko');


axes[6,0].plot(mean,chi_square_pdf(mean,2),'mo',label='Mean');
axes[6,0].plot(median,chi_square_pdf(median,6),'k*',label='Median');
axes[6,0].plot(mode,chi_square_pdf(mode,9),'ks',label='Mode');
axes[6,0].plot(variance,chi_square_pdf(variance,2),'ko',label='Variance');
axes[6,0].axhline(color = "Black");
axes[6,0].axvline(color = "Black");
axes[6,0].set_xlabel("x - axis");
axes[6,0].set_ylabel("Probability Distribution")
axes[6,0].set_title("PDF Of Chi-squared Distribution");
axes[6,0].legend()
    
    
## Cumulative Distribution function of Chi Squared Distribution

    
for i in range(0,len(k)):
    
    x = np.linspace(0,20,1000);
    
    def chi_square_cdf(x,k):
        f = chi2.cdf(x,k);
        return f;
    
    
    axes[6,1].plot(x,chi_square_cdf(x,k[i]),D[i],label=L[i]);
    
axes[6,1].axhline(color = "Black");
axes[6,1].axvline(color = "Black");
axes[6,1].set_xlabel("x - axis");
axes[6,1].set_ylabel("Cumulative Distribution");
axes[6,1].set_title("CDF Of Chi-squared Distribution");
axes[6,1].legend(loc=(0.82,0.05))
    

plt.tight_layout(pad=2);
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random
from mpl_toolkits.mplot3d import Axes3D

%matplotlib notebook
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150

fig = plt.figure()

axes = fig.add_subplot(projection='3d')

mu = 0;
sigma = np.sqrt(4);


x = np.linspace(-5,5);
y = np.linspace(-5,5);
z = np.linspace(0,0);


## Gaussian Distribution function X.


def gaussian_pdf_x(x,mu,sigma):
    f11 = (((x - mu)/sigma)**2)/2;
    f12 = 1/(sigma*((2*np.pi)**0.5));
    f13 = f12 * np.exp(-f11);
    return f13;


## Gaussian Distribution function Y.


def gaussian_pdf_y(y,mu,sigma):
    f21 = (((y - mu)/sigma)**2)/2;
    f22 = 1/(sigma*((2*np.pi)**0.5));
    f23 = f22 * np.exp(-f21);
    return f23;


## Probability Distribution function X^2 + Y^2.


def new_function(x,y,mu,sigma):
    f11 = (((x - mu)/sigma)**2)/2;
    f12 = 1/(sigma*((2*np.pi)**0.5));
    f13 = f12 * np.exp(-f11);
    f21 = (((y - mu)/sigma)**2)/2;
    f22 = 1/(sigma*((2*np.pi)**0.5));
    f23 = f22 * np.exp(-f21);
    f = (f13**2)+(f23**2);
    return f;

## Calculating Mean, Median, Mode and Variance.

A = [];
Z = ((gaussian_pdf_x(x,mu,sigma))**2)+((gaussian_pdf_y(y,mu,sigma))**2);
M = (np.sum(Z))/len(Z);
for i in range(0,len(Z)):
    K = (Z[i] - M)**2
    A.append(K);
    
V = (np.sum(A))/len(A);

## Mean, Median, Mode and Variance of X and Y.

mean = mu;
median = mu;
mode = mu;
variance = (sigma)**2;


## Ploting
axes.plot(x,y,new_function(x,y,mu,sigma),'g',label='X^2 + Y^2');
axes.plot(x,z,gaussian_pdf_x(x,mu,sigma),'r',label='X');
axes.plot(z,y,gaussian_pdf_y(y,mu,sigma),'b',label='Y');
axes.plot(mean,0,gaussian_pdf_x(mean,mu,sigma),'ks',label='Mean of X');
axes.plot(mode,0,gaussian_pdf_x(mode,mu,sigma),'yo',label='Mode of X');
axes.plot(median,0,gaussian_pdf_x(median,mu,sigma),'r.',label='Median of X');
axes.plot(variance,0,gaussian_pdf_x(variance,mu,sigma),'mo',label='Variance of X');
axes.plot(0,mean,gaussian_pdf_y(mean,mu,sigma),'ks',label='Mean of Y');
axes.plot(0,mode,gaussian_pdf_y(mode,mu,sigma),'yo',label='Mode of Y');
axes.plot(0,median,gaussian_pdf_y(median,mu,sigma),'r.',label='Median of Y');
axes.plot(0,variance,gaussian_pdf_y(variance,mu,sigma),'mo',label='Variance of Y');
axes.plot(M,M,new_function(M,M,mu,sigma),'ks',label='Mean of X^2 + Y^2');
axes.plot(M,M,new_function(M,M,mu,sigma),'yo',label='Mode of X^2 + Y^2');
axes.plot(M,M,new_function(M,M,mu,sigma),'r.',label='Median of X^2 + Y^2');
axes.plot(V,V,new_function(V,V,mu,sigma),'k+',label='Variance of X^2 + Y^2');
axes.set_xlabel("x - axis");
axes.set_ylabel("y - axis");
axes.set_zlabel("Probability Distribution");
axes.legend(fontsize=5,loc=(-0.4,0))
plt.tight_layout(pad=2);
```

(a) Consider a transmitter that can transmit messages only in the form of 2 bits 0 and 1. The receiver receives a single message sent by the transmitter say X. Since, we do not know before hand which bit was sent X follows a certain distribution. Which distribution is this? Find the PMF and plot it. Also calculate the mean and variance of X. Assume probability for receiving 0(failure) is p = 0.1, 0.5, 0.8.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig, ax = plt.subplots(1,3,figsize=(16, 4));

p = [0.1,0.5,0.8];
x = range(0,2);
D = ['r.','b.','g.'];

## Bernoulli distribution

def bernoulli_pmf(p,x):
    if x == 0:
        return p;
    elif x == 1:
        return 1-p;
    else:
        return 0;

## Ploting
for i in range(0,len(p)):
    for j in range(0,len(x)):
        ax[i].plot(x[j],bernoulli_pmf(p[i],x[j]),D[i]);
        ax[i].plot(np.linspace(x[j],x[j],1000),np.linspace(0,bernoulli_pmf(p[i],x[j]),1000),D[i]);
        ax[i].axhline(color = "Black");
        ax[i].set_xlabel("x - axis");
        ax[i].set_ylabel("Probability Distribution");
        ax[i].set_xlim(-1,2);
        ax[i].set_ylim(-0.2,1.2);
        
## Calculating Mean and Variance

for i in range(0,len(p)):
    mean = p[i];
    variance = p[i]*(1-p[i]);
    print("Mean at p",i+1," :",mean);
    print("Variance at p",i+1," :",variance);
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline

x = range(0,21);
n = 20;
p = 0.4;
q = 0.6;

## Binomial Distribution

def binomial_pmf(n,x,p,q):
    f1 = np.math.factorial(n)/(np.math.factorial(x)*np.math.factorial(n-x));
    f2 = q**x;
    f3 = p**(n-x);
    f4 = f1*f2*f3;
    return f4;

## Calculating Mean and Varience.

mean = n*p;
variance = n*p*q;
print("Mean :",mean);
print("Variance :",variance);


## Ploting
for i in range(0,len(x)):
    plt.plot(x[i],binomial_pmf(n,x[i],p,q),'go');
    plt.plot(np.linspace(x[i],x[i],1000),np.linspace(0,binomial_pmf(n,x[i],p,q),1000),'g');
plt.xlabel("x - axis");
plt.ylabel("Probability Distribution");
plt.axhline(color = "Black");
plt.axvline(color = "Black");
```
